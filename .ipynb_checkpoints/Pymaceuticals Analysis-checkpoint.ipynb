## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merge_df.head(20)
merge_df.tail(20)
merge_df.head(5)
```

```python
# Checking the number of mice.
mouse_count=merge_df["Mouse ID"].value_counts()
print(f"Current Mouse Count: {len(mouse_count)}")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dupsearch=merge_df[["Mouse ID","Timepoint"]]
dupsearch
dupsearch[merge_df.duplicated(keep=False)]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989_df=merge_df[merge_df["Mouse ID"]=="g989"]
g989_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_merge = merge_df[merge_df['Mouse ID'] != 'g989']
clean_merge
```

```python
# Checking the number of mice in the clean DataFrame.
#print(len(clean_merge))

new_mouse_count=clean_merge["Mouse ID"].value_counts()
print(f"Cleaned Mouse Count: {len(new_mouse_count)}")
```

## Summary Statistics

```python
#find drug names; 
clean_merge[["Drug Regimen","Timepoint","Tumor Volume (mm3)"]]
clean_merge["Drug Regimen"].value_counts()

#change Tumor Volume to int32
clean_merge["Tumor Volume (mm3)"].astype("int32").dtypes

```

```python
# # Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
drug_groups = clean_merge.groupby("Drug Regimen")
#OMG! describe for basic data
summary=drug_groups["Tumor Volume (mm3)"].describe()
summary
```

```python
#adjust summary table to reflect values you want

###MEDIAN
tumor_vol_med = round(drug_groups["Tumor Volume (mm3)"].median(),2)
###SEM
tumor_sem=round(drug_groups['Tumor Volume (mm3)'].sem(),2)
tumor_var=round(drug_groups["Tumor Volume (mm3)"].var(),2)
#add columns you want
summary["Variance"]=tumor_var
summary["Median"]=tumor_vol_med
# summary["Standard Deviation"]=tumor_std
summary["SEM"]=tumor_sem

summary[["mean","std","Median","Variance","SEM"]]
tumor_by_drug_summary=summary[["mean","Median","Variance","std","SEM"]]
tumor_by_drug_summary=tumor_by_drug_summary.rename(columns={"mean":"Mean", "std":"Standard Deviation"})
tumor_by_drug_summary
```

```python
lady_mouse_df["Timepoint"]=lady_mouse_df["Timepoint"].astype(int)
lady_mouse_df["Tumor Volume (mm3)"]=lady_mouse_df["Tumor Volume (mm3)"].astype(int)


timepoints=lady_mouse_df["Timepoint"]
tumor_volume=lady_mouse_df["Tumor Volume (mm3)"]


x_values=timepoints
y_values=tumor_volume
plt.plot(timepoints, tumor_volume)
plt.title("Capomulin: Affect on Tumor Size for Mouse s185")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Size (mm3)")

```

```python
###gathering entries for all mice treated with Capomulin, pulling out relevant columns
capo_study=clean_merge[(clean_merge["Drug Regimen"]=="Capomulin")]
capo_study=capo_study[["Drug Regimen", "Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]
average_data=capo_study.groupby("Mouse ID").mean()
average_data.head(5)
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
av_tum_vol=average_data["Tumor Volume (mm3)"]
mouse_weight=average_data["Weight (g)"]


plt.scatter(mouse_weight,av_tum_vol, marker='o', facecolors='m', edgecolors='black')

plt.title("Average Tumor Size Compared to Weight for\n Mice Treated With Capomulin")

plt.xlabel("weight in (g)")
plt.ylabel("tumor size in (mm)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
cc=round(st.pearsonr(mouse_weight, av_tum_vol)[0],2)
print(f"Correlation Coefficient: {cc}")
```

```python
vc_slope, vc_int, vc_r, vc_p, vc_std_err = st.linregress(mouse_weight, av_tum_vol)
vc_fit = vc_slope * mouse_weight + vc_int
```

```python
plt.scatter(mouse_weight,av_tum_vol, marker='o', facecolors='g', edgecolors='g')
plt.plot(mouse_weight,vc_fit,"-")

plt.title("Average Tumor Size Compared to Weight for\n Mice Treated With Capomulin")

plt.xlabel("weight in (g)")
plt.ylabel("tumor size in (mm)")
plt.show()
```
